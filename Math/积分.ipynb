```python
def square(x):
    return 2*x**2 + 3*x +4
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
X = np.linspace(-10,10,1000)
```

```python
Y = [square(x) for x in X]
```

```python
print(Y)
```

```python
plt.plot(X,Y)
```

```python
def get_the_area_apporaximate(f, Xs):
    area = 0
    for i,x in enumerate(Xs[:-1]):
        next_x = Xs[i+1]
        delta_x = next_x - x
        sub_area = ((f(x)+f(next_x))*delta_x)/2
        area += sub_area
    return area
```

```python
get_the_area_apporaximate(square,np.linspace(-10,10,1))
```

```python
get_the_area_apporaximate(square,np.linspace(-10,10,3
                                            ))
```

```python
get_the_area_apporaximate(square,np.linspace(-10,10,1000))
```

```python
get_the_area_apporaximate(square,np.linspace(-10,10,10000))
```
